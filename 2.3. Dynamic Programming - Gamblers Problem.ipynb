```julia
using PyPlot
```

```julia
list_of_possible_actions(s) = [1:minimum([s,100-s])]

#println("possible bets if s = 70: $(list_of_possible_actions(70))")
#println("possible bets if s = 10: $(list_of_possible_actions(10))")
```

```julia
function backup_state(s::Int64, a::Int64)
    pw = 0.4
    
    r = pw * V[s+a]
    if s - a > 0
     r = r + (1-pw) * V[s-a]
    end
    
    return r
end

#expected_return_term(90, 10, 0.9, 0.4)
```

```julia
function get_policy(s::Int64)
    possible_actions = list_of_possible_actions(s)
    action           = possible_actions[1]
    best             = backup_state(s, action)
    possible_actions = possible_actions[2:end]
    for a in possible_actions
        current = backup_state(s, a)
        if current > best
            best = current
            action = a
        end
    end
    return action
end

#get_policy(51)
```

```julia
function one_step!(V::Vector{Float64})
    old_values = copy(V)
    for s = 1:99
        V[s] = maximum([backup_state(s, a) for a in list_of_possible_actions(s)])
    end
    return maximum(abs(old_values .- V))
end

function full_iteration(V::Vector{Float64}, θ::Float64)
    δ = θ + 0.1
    while δ > θ
        δ = one_step!(V)
    end
    return δ
end

#one_step!(V, 0.9, 0.3, 0.1)
```

```julia
function plot_results()
    figure(figsize=(10,3))
    subplot(121)
    plot(V[1:100])
    subplot(122)
    policy = [get_policy(s) for s=1:99]
    plot(policy)
end

#plot_results(V, 0.9, 0.35)
```
